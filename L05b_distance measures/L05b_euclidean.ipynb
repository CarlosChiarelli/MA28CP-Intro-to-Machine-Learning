```
import numpy as np

x1 = np.array([1,2])
x2 = np.array([3,1]) 
```

```
def emat(p,q):
    import math
    return np.sqrt(math.pow((p[0] - q[0]), 2) + math.pow((p[1] - q[1]), 2))

emat(x1,x2)
```

$d(\mathbf{p},\mathbf{q}) = \sqrt{\sum_{i=1}^n (p_i-q_i)^2}$

```
def sqrt_sum(p,q):
    return np.sqrt(np.sum((p-q)**2, axis=0))

sqrt_sum(x1,x2)
```

$\mathbf{a} = \mathbf{p} -\mathbf{q} = (q_{1}-p_{1},q_{2}-p_{2},\cdots ,q_{n}-p_{n})$

$\mathbf{a} = (1-3, 2-1) = (-2,1)$

$\sqrt{\mathbf {a}^T \mathbf {a}} =\sqrt{{\begin{bmatrix}a_{1}&a_{2}&...&a_{n}\end{bmatrix}}{\begin{bmatrix}a_{1}\\a_{2}\\...\\a_{n}\end{bmatrix}}}=\sqrt{a_{1} a_{1}+a_{2} a_{2}+...+a_{n} a_{n}}=\sqrt{\sum _{i=1}^{n}{\bigl (}{a_{i}}^{2}{\bigr )}}$

```
def bymatmult(p,q):
    return np.sqrt((p-q) @ (p-q))

bymatmult(x1,x2)
```

```
def bymatmult2(p,q):
    return np.sqrt(np.matmul(np.transpose(p-q),p-q))

bymatmult2(x1,x2)
```

$\left\|{\mathbf  {q}}-{\mathbf  {p}}\right\|={\sqrt  {({\mathbf  {q}}-{\mathbf  {p}})\cdot ({\mathbf  {q}}-{\mathbf  {p}})}}$

$\left\|\mathbf {a} \right\|={\sqrt {\mathbf {a} \cdot \mathbf {a} }}={\sqrt {a_{1}^{2}+a_{2}^{2}+\cdots +a_{n}^{2}}}$


```
def linalg_norm(p,q):
    return np.linalg.norm(p-q)

linalg_norm(x1,x2)
```

$\left\|{\mathbf  {a}}\right\|={\sqrt  {a_{1}^{2}+a_{2}^{2}+\cdots +a_{n}^{2}}}={\sqrt  {{\mathbf  {a}}\cdot {\mathbf  {a}}}}$

```
def edot(p,q):
    return np.sqrt(np.dot(p-q,p-q))

edot (x1,x2)
```

```
def scipy_distance(p,q):
    from scipy.spatial import distance
    return distance.euclidean(p,q)

scipy_distance(x1,x2)
```
