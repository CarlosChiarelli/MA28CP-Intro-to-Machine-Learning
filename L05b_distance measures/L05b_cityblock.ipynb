```
import numpy as np

x1 = np.array([1,2])
x2 = np.array([3,1]) 
```

$d(\mathbf{p},\mathbf{q}) = (\sum_{i=1}^n |p_i-q_i|^1)^{(1/1)}$

$d(\mathbf{p},\mathbf{q}) = (|p_1-q_1|^1 + |p_2-q_2|^1)^{(1/1)}$

para $\mathbf{p}=[1,2]$ e $\mathbf{q}=[3,1]$

$d(\mathbf{p},\mathbf{q}) = (|1-3|^1 + |2-1|^1)^{(1/1)}$

$d(\mathbf{p},\mathbf{q}) = (|-2|^1 + |1|^1)^{(1/1)}$

$d(\mathbf{p},\mathbf{q}) = (2^1 + 1^1)^{(1/1)}$

$d(\mathbf{p},\mathbf{q}) = (2 + 1)^{(1/1)}$

$d(\mathbf{p},\mathbf{q}) = 3^{(1/1)}$

$d(\mathbf{p},\mathbf{q}) = 3$

```
def emat(p,q):
    return pow(pow(abs(p[0] - q[0]), 1) + pow(abs(p[1] - q[1]), 1), 1/1)

emat(x1,x2)
```

$d(\mathbf{p},\mathbf{q}) = (\sum_{i=1}^n |p_i-q_i|^1)^{(1/1)}$

```
def sqrt_sum(p,q):
    return pow(np.sum(abs(p-q)**1, axis=0), 1/1)

sqrt_sum(x1,x2)
```

$\mathbf{a} = \mathbf{p} -\mathbf{q} = (q_{1}-p_{1},q_{2}-p_{2},\cdots ,q_{n}-p_{n})$

$\mathbf{a} = (1-3, 2-1) = (-2,1)$

$\left\|{\mathbf {a}}\right\|_{1}= \sum_{i=1}^{n}\left|a_{i}\right| $

$\left\|{\mathbf {a}}\right\|_{1} = |a_{1}|+|a_{2}|+\cdots +|a_{n}|$

```
def edot(p,q):
    return np.sum(abs(p-q))

edot (x1,x2)
```

```
def scipy_distance(p,q):
    from scipy.spatial import distance
    return distance.cityblock(p,q)

scipy_distance(x1,x2)
```
