```
import numpy as np

x1 = np.array([1,2])
x2 = np.array([3,1]) 
```

$d(\mathbf{p},\mathbf{q}) = \max _{i}(|p_{i}-q_{i}|)$

```
def sqrt_sum(p,q):
    return max(abs(p-q))

sqrt_sum(x1,x2)
```

$\mathbf{a} = \mathbf{p} -\mathbf{q} = (q_{1}-p_{1},q_{2}-p_{2},\cdots ,q_{n}-p_{n})$

$\mathbf{a} = (1-3, 2-1) = (-2,1)$

$\left\|\mathbf {a} \right\|_{\infty }:=\max _{i}(\left|a_{i}\right|)$

$\left\|\mathbf {a} \right\|_{\infty }:=\max_i \left(\left|a_{1}\right|,\ldots ,\left|a_{n}\right|\right)$

```
def scipy_distance(p,q):
    from scipy.spatial import distance
    return distance.chebyshev(p,q)

scipy_distance(x1,x2)
```
